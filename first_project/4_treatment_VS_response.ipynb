```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
cancer_df= pd.read_csv("cancerDataSet\cancer issue.csv")
```

```python
plt.figure(figsize=(10, 6))

# Create a bar plot of TreatmentType vs TreatmentResponse
sns.countplot(data=cancer_df, x='TreatmentType', hue='TreatmentResponse', palette='Blues')

# Enhance plot appearance
plt.title('Distribution of Treatment Response by Treatment Type', fontsize=16, fontweight='bold')
plt.xlabel('Treatment Type', fontsize=14)
plt.ylabel('Count of Responses', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Treatment Response', fontsize=12)
plt.ylim(400,1800)
plt.tight_layout()

# Show the plot
plt.show()
```
