```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
cancer_df= pd.read_csv("cancerDataSet\cancer issue.csv")
```

```python
stage_stats = cancer_df.groupby("Stage")["SurvivalMonths"].mean()
stage_stats
```

```python
type_stats = cancer_df.groupby("CancerType")["SurvivalMonths"].mean()
type_stats
```

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 6))

# Plot stage statistics as a horizontal bar chart
stage_stats.sort_values().plot(kind='barh', ax=axes[0])
axes[0].set_title("Mean Survival Months by Cancer Stage")
axes[0].set_xlabel("Mean Survival Months")
axes[0].set_ylabel("Cancer Stage")
axes[0].set_xlim(55,65)


# Plot type statistics as a horizontal bar chart
type_stats.sort_values().plot(kind='barh', ax=axes[1])
axes[1].set_title("Mean Survival Months by Cancer Type")
axes[1].set_xlabel("Mean Survival Months")
axes[1].set_ylabel("Cancer Type")
axes[1].set_xlim(55,65)


# Adjust layout
plt.tight_layout()
```
