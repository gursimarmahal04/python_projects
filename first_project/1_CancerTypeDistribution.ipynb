```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
cancer_df= pd.read_csv("cancerDataSet\cancer issue.csv")
```

```python
cancerType= cancer_df.CancerType.value_counts()
cancerType.plot(kind='barh')

plt.title("Most common cancer type", fontsize=16)
plt.ylabel("Cancer Type", fontsize=12)
plt.xlabel("counts", fontsize=12)
plt.xlim(2800,3000)
ax=plt.gca()
ax.invert_yaxis()
plt.tight_layout()
```

```python
sorted_cancer_types = cancer_df.groupby("CancerType")["Age"].median().sort_values(ascending=False).index
```

```python
sns.boxplot(x="CancerType", y="Age", data=cancer_df, order=sorted_cancer_types, palette="Blues")

# Add labels and title
plt.title("Age Distribution by Cancer Type", fontsize=16)
plt.xlabel("Cancer Type", fontsize=12)
plt.ylabel("Age", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
```
