```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
cancer_df= pd.read_csv("cancerDataSet\cancer issue.csv")
```

```python
cancer_df.groupby('TumorSize')['SurvivalMonths'].median().reset_index().plot(
    kind='line',
    x='TumorSize',
    y='SurvivalMonths',
    marker='o',  # Adds markers to the data points
    color='royalblue',  # Custom color for the line
    linewidth=2,  # Thicker line
    markersize=6,  # Larger markers
    linestyle='-',  # Solid line
    grid=True,  # Show gridlines
    figsize=(10, 6)  # Bigger figure for better visibility
)

# Enhancing plot appearance
plt.title('Average Survival Months by Tumor Size', fontsize=16, fontweight='bold', color='darkslategray')
plt.xlabel('Tumor Size', fontsize=14, color='dimgray')
plt.ylabel('Average Survival Months', fontsize=14, color='dimgray')
plt.tick_params(axis='both', labelsize=12)
plt.grid(True, linestyle='--', alpha=0.5)  # Grid with dashed lines

# Show the plot
plt.tight_layout()
plt.show()
```
